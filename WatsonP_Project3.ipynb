💚 Background

In short, data leakage in machine learning is a term used to describe a case where the data used to train an algorithm includes unexpected additional information about the subject it’s evaluating.

How Does Data Leakage Happen?
Data leakage in machine learning can happen in various ways during the data handling and preparation stage. For instance, if you scale or normalize the entire data set before splitting it, you risk unintentionally mixing information. Another trouble spot is feature engineering. Creating new features from the complete data set before its division can embed insights from the test data into the training data, potentially leading to data leakage. You should also be aware of improper data splitting, where data is not accurately divided during training and testing. Similarly, using unverified external sources can introduce forward-looking information, compromising the model’s integrity. It’s possible to inadvertently induce data leakage at any stage, including validation and process changes.

#### Precision and Recall

The precision of a model measures how many of a model's predictions (prediction = 1) are legitimate (labelCol (Target) = 1). The recall of a model shows how many of the true positives (Target = 1) are caught by the model (my_prediction = 1). The F1 score can be interpreted as a harmonic mean of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0. The relative contribution of precision and recall to the F1 score are equal.

Precision = TP / (TP + FP)

Recall = TP / (TP + FN)

F1 score = (2 X Precision X Recall) / (Precision + Recall)

AVERAGE F1 =

[#rowsF(0) / totalrows] X (F1(0)] + [#rowsF(1) / total rows] X (F1(1)]

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html

## Model 3: Decision Tree Classifier, same input features as Model 2

A decision tree classifier works by recursively splitting the input data into subsets based on the values of input features. The splits are chosen to maximize the homogeneity of the target variable within the subsets, typically using measures like Gini impurity or entropy.

Starting from the root node, which contains all the data, the algorithm selects the feature and split point that best separates the data into two subsets. This process is repeated for each subset, creating a tree-like structure where each internal node represents a feature and a split point, and each leaf node represents a class label or a probability distribution over the classes.

To classify a new data point, the algorithm follows the decisions at each internal node until it reaches a leaf node, which gives the predicted class label.

## Model 4: Random Forest Classifier

Random Forest is a popular machine learning algorithm used for both classification and regression tasks. It is a type of ensemble learning method, which means it combines multiple decision trees to make a more accurate prediction.

The algorithm works by creating multiple decision trees, each trained on a different subset of the data. During the prediction phase, each decision tree makes a prediction and the final output is the average or majority vote of all the decision trees. This combination of multiple decision trees helps to reduce overfitting and improve the overall accuracy of the model.

One of the advantages of Random Forest is its ability to handle large amounts of data and a high number of features. It also has the ability to identify important features in the data, which can be useful for feature selection. Additionally, it is less prone to overfitting than a single decision tree, which makes it a more robust model.

Another advantage of Random Forest is that it can be used for both classification and regression tasks. In classification, the algorithm predicts a class label, while in regression it predicts a numerical value.

However, one of the disadvantages of Random Forest is that it can be computationally expensive, especially when the number of decision trees is high. Additionally, it can be difficult to interpret the results of a Random Forest model, as it is a black box model.

As a result, the naive Bayes classifier is a powerful tool in machine learning, particularly in text classification, spam filtering, and sentiment analysis, among others.

### Optimizing Hyperparameters with Cross-Validation

By fine-tuning some aspects of the model training (how Spark builds the
fitted model), we can hope to yield better model accuracy. For this, we use a technique called cross-validation. Cross-validation resamples the data set into training and testing sets to assess the ability of the model to generalize over new data. After identifying the best ParamMap, CrossValidator finally re-fits the Estimator using the best ParamMap and the entire dataset.

CrossValidator begins by splitting the dataset into a set of folds which are used as separate training and test datasets. E.g., with k=3
 folds, CrossValidator will generate 3 (training, test) dataset pairs, each of which uses 2/3 of the data for training and 1/3 for testing. To evaluate a particular ParamMap, CrossValidator computes the average evaluation metric for the 3 Models produced by fitting the Estimator on the 3 different (training, test) dataset pairs.

After identifying the best ParamMap, CrossValidator finally re-fits the Estimator using the best ParamMap and the entire dataset.

https://spark.apache.org/docs/2.2.0/ml-tuning.html

    NO CHANGE AFTER LIMITING FEATURES WHERE FEATURE_IMPORTANCE > 0

    There are a couple of reasons why the weighted F1 score might remain the same after reducing features based on importance in a decision tree classifier:

1. Redundant or Correlated Features:

If the features you removed have high importance but are highly correlated or redundant with remaining features, removing them might not significantly impact the model's ability to distinguish between classes. The remaining features might already capture the information needed for classification.
2. Threshold Choice for Feature Importance:

The chosen threshold for feature importance (feature_importance > 0) might be too low. Features with slightly lower importance might still contribute to the model's performance, especially in interaction with other features. Experiment with different thresholds to see if a higher cutoff improves the F1 score.
3. Limited Information in High-Importance Features:

It's possible that the features with high importance by themselves don't provide enough information for accurate classification. The model might rely on interactions between features with lower importance to make correct predictions. Removing these features could disrupt these interactions.
4. Nature of the Classification Problem:

In some classification problems, a small number of highly informative features might be sufficient for good performance. In such cases, removing features with lower importance might not significantly affect the F1 score.
Here's what you can do to investigate further:

Analyze Removed Features: Look at the features you removed and see if they are highly correlated with remaining ones. Consider domain knowledge to understand their potential role in classification.
Experiment with Thresholds: Try different thresholds for feature importance selection and see if a stricter cutoff (e.g., feature_importance > 0.1) improves the F1 score.

💚 A Note About Chi-Square and p-Values

In a chi-square test, the p-value and the test statistic are two important results that help you determine the significance of the relationship between categorical variables in your data. Here's how to interpret them:

p-value: This is a measure of the probability that the observed relationship between the variables is due to chance. A low p-value (usually below 0.05) indicates that the relationship is statistically significant, meaning that it is unlikely to have occurred by random chance alone. In other words, if the p-value is low, you can reject the null hypothesis and conclude that there is a significant association between the variables.

Chi-square statistic: This is a measure of how much the observed frequencies of the categorical variables deviate from the expected frequencies under the null hypothesis of no association. A larger chi-square statistic indicates a greater deviation, which suggests a stronger association between the variables.

When interpreting the results of a chi-square test, you typically look at both the p-value and the chi-square statistic. If the p-value is low and the chi-square statistic is large, you can conclude that there is a significant association between the variables. If the p-value is high and/or the chi-square statistic is small, you fail to reject the null hypothesis, suggesting that there is no significant association.
